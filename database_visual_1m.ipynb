```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets_1 = pd.read_csv('./data/[1m]BTCUSDT.csv')
```

```python
datasets_1 
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets_1['Start'], datasets_1['Open'], color = 'black')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets_1['Stop'], datasets_1['Close'], color = 'blue')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets_1['Start'], datasets_1['Open'], 'black')
plt.plot(datasets_1['Stop'], datasets_1['Close'], 'orange')
```

```python
plt.figure(figsize=(30, 10))
plt.plot(datasets_1['Start'], datasets_1['Low'], color = 'black')
plt.plot(datasets_1['Start'], datasets_1['High'], color = 'green')
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets_1['Start'] + datasets_1['Stop']) / 2, datasets_1['Volume'], 'orange', datasets_1['Start'], datasets_1['Open'], 'black')
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets_1['Start'] + datasets_1['Stop'])/2, datasets_1['Volume'], 'violet', linestyle="",marker="o", alpha=0.15)
```

```python
plt.figure(figsize=(30, 9))
plt.plot(abs(datasets_1['High'] - datasets_1['Low']), datasets_1['Volume'], 'blue', linestyle="",marker="o", alpha=0.15)
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets_1['Start'] + datasets_1['Stop'])/2, abs(datasets_1['High'] - datasets_1['Low']), 'coral', linestyle="",marker="o", alpha=0.15)
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets_1['Start'] + datasets_1['Stop'])/2, abs(datasets_1['High'] - datasets_1['Low']), 'orange')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets_1['High'] - datasets_1['Low'], 'green')
```
