```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime

def dateparse(time):
    return datetime.datetime.fromtimestamp(float(time)/1000)

data = pd.read_csv('/home/maksym/2021-knu-market/data/[1h]BTCUSDT.csv', parse_dates=['Start', 'Stop'],date_parser=dateparse)

data.head()
```

```python
#Open-Close price plot

plt.figure(figsize=(20, 10))
plt.plot(data.Start, data.Open, 'b', label='Open price')
plt.plot(data.Stop, data.Close, 'r', label='Close price')
plt.legend(fontsize=15)
```

```python
#Low and High price plot

plt.figure(figsize=(20, 10))
plt.plot(data.Stop, data.Low, 'b', label='Low price')
plt.plot(data.Stop, data.High, 'r', label='High price')
plt.legend(fontsize=15)

```

```python
#volume and mean price per hour plot

mean_price_per_hour = (data['High'].copy() + data['Low'].copy())/2

plt.figure(figsize=(20, 10))
plt.plot(data.Stop, data.Volume, 'b', label='Volume')
plt.plot(data.Stop, mean_price_per_hour, 'r', label='Mean price per hour')
plt.legend(fontsize=15)

```

```python
#scatter matrix
sm = pd.plotting.scatter_matrix(data[['Open', 'Close', 'Low', 'High', 'Volume']], marker='o',figsize=(15,15))
```
