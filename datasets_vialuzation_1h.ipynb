```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets = pd.read_csv('./data/[1h]BTCUSDT.csv')
```

```python
datasets
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets['Start'], datasets['Open'], color = 'black')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets['Stop'], datasets['Close'], color = 'black')
```

```python
plt.figure(figsize=(30, 10))
plt.plot(datasets['Start'], datasets['Open'], color = 'black')
plt.plot(datasets['Stop'], datasets['Close'], color = 'orange')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets['Start'], datasets['Low'], color = 'black')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets['Start'], datasets['High'], color = 'black')
```

```python
plt.figure(figsize=(30, 10))
plt.plot(datasets['Start'], datasets['Low'], color = 'black')
plt.plot(datasets['Start'], datasets['High'], color = 'orange')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets['Start'], datasets['Volume'], color = 'black')
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets['Start'] + datasets['Stop']) / 2, datasets['Volume'], 'orange', datasets['Start'], datasets['Open'], 'black')
```

```python
plt.figure(figsize=(30, 9))
plt.plot(datasets['High'] - datasets['Low'], 'green')
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets['Start'] + datasets['Stop'])/2, abs(datasets['High'] - datasets['Low']), 'orange')
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets['Start'] + datasets['Stop'])/2, datasets['Volume'], 'violet', linestyle="",marker="o", alpha=0.15)
```

```python
plt.figure(figsize=(30, 9))
plt.plot(abs(datasets['High'] - datasets['Low']), datasets['Volume'], 'blue', linestyle="",marker="o", alpha=0.15)
```

```python
plt.figure(figsize=(30, 9))
plt.plot((datasets['Start'] + datasets['Stop'])/2, abs(datasets['High'] - datasets['Low']), 'coral', linestyle="",marker="o", alpha=0.15)
```
