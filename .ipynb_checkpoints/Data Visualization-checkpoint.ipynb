```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# datasetH

```python
datasetH = pd.read_csv("./data/[1h]BTCUSDT.csv")
```

```python
datasetH
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Open'], color = 'red')
plt.title("Open price", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Stop'], datasetH['Close'], color = 'red')
plt.title("Close price", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Open'], c='red', label="open price")
plt.plot(datasetH['Stop'], datasetH['Close'], c='black', label="cloce price")
plt.title("Open and Close price", fontsize=10)
plt.legend()
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Low'], color = 'blue')
plt.title("Low price", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['High'], color = 'red')
plt.title("High price", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Low'], color = 'blue', label='low price')
plt.plot(datasetH['Start'], datasetH['High'], color = 'red', label='high price')
plt.title("High and Low price", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Volume'], color = 'gray')
plt.title("Volume", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['High'] - datasetH['Low'], 'yellow')
plt.title("High price - Low price", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot((datasetH['Start'] + datasetH['Stop']) / 2, datasetH['Volume'], 'gray', datasetH['Start'], datasetH['Open'], 'black')
plt.title("Open price and Volume", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.hist(datasetH['Volume'])
plt.title("Volume", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
sns.histplot(datasetH, x="Volume")
plt.title("Volume", fontsize=10)
```

```python
plt.figure(figsize=(20, 5))
plt.plot((datasetH['High'] - datasetH['Low']), datasetH['Volume'], 'red', linestyle="",marker="o", alpha=0.1)
plt.title("High price - Low Price by Volume", fontsize=10)
```
