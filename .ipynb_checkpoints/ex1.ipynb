```python
import matplotlib.pyplot as mpl
import pandas as pd
import seaborn as sb
dt = pd.read_csv("[1h]BTCUSDT.csv")
dt
```

```python
mpl.figure(figsize=(17,13))
mpl.plot(dt['Start'], dt['Open'], c='green')
mpl.plot(dt['Stop'], dt['Close'], c='red')
mpl.show()
```

```python
mpl.figure(figsize=(17, 13))
mpl.plot(dt['Start'], dt['High'], c='green')
mpl.plot(dt['Stop'], dt['Low'], c='red')
mpl.show()
```

```python
mpl.figure(figsize=(17,13))
mpl.plot(dt['Start'], dt['Volume'], c='green')
```

```python
mpl.hist(dt['Volume'], color='black')
```

```python
mpl.hist(dt['Volume'],bins=50)
```
