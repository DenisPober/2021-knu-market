```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('./data/[1h]BTCUSDT.csv')
data_2 = pd.read_csv('./data/[1m]BTCUSDT.csv')
```

```python
x_h = pd.to_datetime(data_1['Stop'].to_numpy(), unit='ms').to_numpy()
y_h = data_1['Close'].to_numpy()
```

```python
plt.figure(figsize=(30, 9))
plt.plot(x_h, y_h)
```

```python
x_m = pd.to_datetime(data_2['Stop'].to_numpy(), unit='ms').to_numpy()
y_m = data_2['Close'].to_numpy()
```

```python
plt.figure(figsize=(30, 9))
plt.plot(x_m, y_m)
```

```python
#Autoregression (AR) (Data_1_h)
from statsmodels.tsa.ar_model import AutoReg
```

```python
model_h = AutoReg(y_h, lags=1)
model_fit = model_h.fit()
yhat_h = model_fit.predict(0, len(y_h) + 1000)
print(yhat_h)
```

```python
plt.figure(figsize=(30, 9))
plt.plot(range(len(yhat_h)), yhat_h, color='black')
plt.plot(range(len(y_h)), y_h, c='violet')
plt.show()
```

```python
#Autoregression (AR) (Data_2_m)
model_m = AutoReg(y_m, lags=1)
model_fit = model_m.fit()
yhat_m = model_fit.predict(0, len(y_m) + 100)
print(yhat_m)
```

```python
plt.figure(figsize=(30,9))
plt.plot(range(len(yhat_m)), yhat_m, color='black')
plt.plot(range(len(y_m)), y_m, c='coral')
plt.show()
```

```python
#Moving Average (MA) (Data_1_h)
from statsmodels.tsa.arima.model import ARIMA
```

```python
model_h = ARIMA(y_h, order=(0, 0, 1))
model_fit = model_h.fit()
yhat_h = model_fit.predict(0, len(y_h)+1000)
print(yhat_h)
```

```python
plt.figure(figsize=(30,9))
plt.plot(range(len(yhat_h)), yhat_h, c='violet')
plt.plot(range(len(y_h)), y_h, c='black')
plt.show()
```

```python
#Simple Exponential Smoothing (SES) (data_1_h)
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
model_h = SimpleExpSmoothing(y_h)
model_fit = model_h.fit()
yhat_h = model_fit.predict(0, len(y_h)+ 10000)
print(yhat_h)
```

```python
plt.figure(figsize=(30,9))
plt.plot(range(len(yhat_h)), yhat_h, c='black')
plt.plot(range(len(y_h)), y_h, c='violet')
plt.show()
```

```python
#Simple Exponential Smoothing (SES) (data_1_m)
model_m = SimpleExpSmoothing(y_m)
model_fit = model_m.fit()
yhat_m = model_fit.predict(0, len(y_m)+ 100000)
print(yhat_m)
```

```python
plt.figure(figsize=(30,9))
plt.plot(range(len(yhat_m)), yhat_m, c='black')
plt.plot(range(len(y_m)), y_m, c='coral')
plt.show()
```

```python
#Holt Winter’s Exponential Smoothing (HWES) (data_1_h)
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```
