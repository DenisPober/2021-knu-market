```python
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/[1h]BTCUSDT.csv')
```

```python
index = [x for x in range(0, len(df["Open"]))]
```

```python
fig, ax = plt.subplots()
line1, = ax.plot(index, df["Close"],label="Time/Open price")
line2, = ax.plot(index, df["Open"],label="Time/Close price")
line3, = ax.plot(index, df["Volume"],label="Time/Volume")
ax.set(xlabel='Time', ylabel='CLose price')
ax.legend()
plt.show()
```
