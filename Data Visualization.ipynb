```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datasetH = pd.read_csv("./data/[1h]BTCUSDT.csv")
```

```python
datasetH
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Open'], color = 'red')
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Stop'], datasetH['Close'], color = 'red')
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Open'], c='red')
plt.plot(datasetH['Stop'], datasetH['Close'], c='black')
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Low'], color = 'blue')
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['High'], color = 'red')
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Low'], color = 'blue')
plt.plot(datasetH['Start'], datasetH['High'], color = 'red')
```

```python
plt.figure(figsize=(20, 5))
plt.plot(datasetH['Start'], datasetH['Volume'], color = 'gray')
```

```python
plt.figure(figsize=(20, 5))
plt.plot((datasetH['Start'] + datasetH['Stop']) / 2, datasetH['Volume'], 'gray', datasetH['Start'], datasetH['Open'], 'black')
```

```python
plt.figure(figsize=(20, 5))
plt.hist(datasetH['Volume'])
```

```python
plt.figure(figsize=(20, 5))
sns.histplot(datasetH, x="Volume")
```
